```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load the data from the CSV file
data = pd.read_csv('MATH 547 Introductions.csv')

# Get the shape of the dataset (number of rows and columns)
shape = data.shape
print(f'The dataset has {shape[0]} rows and {shape[1]} columns.')
print(data.head())
```

```python
data_numeric = data.select_dtypes(include=[np.number])

# Calculate the rank of the data matrix
rank = np.linalg.matrix_rank(data_numeric)
print(rank)
```

```python

# Extract the specific column for Machine Learning excitement
ml_column = 'In this section, please indicate how excited you are by the following topics. 1 indicates "not excited" and 5 indicates "very excited" [Machine Learning]'

# Calculate the average excitement for Machine Learning
average_ml = data[ml_column].mean()

# Extract the specific column for Machine Learning excitement
svd_column = 'In this section, please indicate how excited you are by the following topics. 1 indicates "not excited" and 5 indicates "very excited" [Singular Value Decomposition]'

# Calculate the average excitement for Machine Learning
average_ml = data[ml_column].mean()
average_svd = data[svd_column].mean()
# Display the result
print(f"Average excitement for Machine Learning: {average_ml:.2f}")
print(f"Average excitement for SVD: {average_svd:.2f}")

rd_column = 'In this section, please indicate how excited you are by the following topics. 1 indicates "not excited" and 5 indicates "very excited" [Working with Real Data]'
average_rd = data[rd_column].mean()

prog_column = 'In this section, please indicate how excited you are by the following topics. 1 indicates "not excited" and 5 indicates "very excited" [Programming (MATLAB/Python)]'
average_prog = data[prog_column].mean()
print(f"Average excitement for working with real data: {average_rd:.2f}")
print(f"Average excitement for programming: {average_prog:.2f}")
```

```python
background = data['Choose the best description of your background.']
ml_excitement = data[ml_column]
svd_excitement = data[svd_column]
rd_excitement = data[rd_column]
prog_excitement = data[prog_column]
interests_df = pd.DataFrame({
    'Background': background,
    'Machine Learning': ml_excitement,
    'Singular Value Decomposition': svd_excitement,
    'Working with Real Data': rd_excitement,
    'Programming': prog_excitement
})

# Group the data by background and calculate the mean excitement for each background group
grouped = interests_df.groupby('Background').mean()

# Plot the relationship
plt.figure(figsize=(10, 6))

# Plot each academic interest (column) vs the background
for column in grouped.columns:
    plt.plot(grouped.index, grouped[column], label=column, marker='o')

# Add labels and title
plt.title('Relationship between Students’ Interests and Skills', fontsize=16)
plt.xlabel('Background', fontsize=14)
plt.ylabel('Average Excitement Level', fontsize=14)
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
data_numerical = data.select_dtypes(include=[np.number])

U, S, Vt = np.linalg.svd(data_numerical)

# The shapes of U, S, and V
print(f"Shape of U: {U.shape}")
print(f"Shape of S (Sigma): {S.shape}")
print(f"Shape of V^T: {Vt.shape}")
```

```python
# Plot the singular values (S)
plt.figure(figsize=(8, 6))
plt.plot(S, marker='o', linestyle='-', color='b')
plt.title('Scree Plot of Singular Values')
plt.xlabel('Index')
plt.ylabel('Singular Value')
plt.grid(True)
plt.show()
```

```python
data_matrix = data_numerical
# Compute ATA and AAT
ATA = np.dot(data_matrix.T, data_matrix)
AAT = np.dot(data_matrix, data_matrix.T)

# Plot ATA
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(ATA, cmap='viridis', aspect='auto')
plt.title('Heatmap of ATA')
plt.colorbar()

# Plot AAT
plt.subplot(1, 2, 2)
plt.imshow(AAT, cmap='viridis', aspect='auto')
plt.title('Heatmap of AAT')
plt.colorbar()

plt.tight_layout()
plt.show()

# Print the shapes
print(f"Shape of ATA: {ATA.shape}")
print(f"Shape of AAT: {AAT.shape}")
```

```python
# Compute the SVD of ATA and AAT
U_ATA, S_ATA, Vt_ATA = np.linalg.svd(ATA)
U_AAT, S_AAT, Vt_AAT = np.linalg.svd(AAT)

# Plot the singular values for ATA and AAT
plt.figure(figsize=(8, 6))

plt.subplot(1, 2, 1)
plt.plot(S_ATA, marker='o', linestyle='-', color='g')
plt.title('Singular Values of ATA')
plt.xlabel('Index')
plt.ylabel('Singular Value')

plt.subplot(1, 2, 2)
plt.plot(S_AAT, marker='o', linestyle='-', color='r')
plt.title('Singular Values of AAT')
plt.xlabel('Index')
plt.ylabel('Singular Value')

plt.tight_layout()
plt.show()

# Compare the singular values of A, ATA, and AAT
print(f"Singular values of A: {S}")
print(f"Singular values of ATA: {S_ATA}")
print(f"Singular values of AAT: {S_AAT}")
```
